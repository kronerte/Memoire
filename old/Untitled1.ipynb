```python
def q_q(X,Y):
    X_q = np.zeros(100)
    Y_q = np.zeros(100)
    for i in range(100):
        X_q[i] = np.quantile(X, i/100)
        Y_q[i] = np.quantile(Y, i/100)
    return X_q, Y_q
```

```python
X = np.random.randn(1000000)
Y = np.random.randn(1000000)
X_q, Y_q = q_q(X,Y)
plt.scatter(X_q, Y_q)
x = np.linspace(-5,5,100)
f = lambda x:x
plt.plot(x,f(x), '--')
plt.xlabel("Normal distribution quantile")
plt.ylabel("Mesured distribution quantile")
plt.savefig("QQplotnormalnormal.eps")
```

```python
X = np.random.randn(1000000)
Y = np.random.laplace(size=1000000)
X_q, Y_q = q_q(X,Y)
plt.scatter(X_q, Y_q)
x = np.linspace(-5,5,100)
f = lambda x:x
plt.plot(x,f(x), '--')
plt.xlabel("Normal distribution quantile")
plt.ylabel("Mesured distribution quantile")
plt.savefig("QQplotlaplacenormal.eps")
```

```python
X = np.random.randn(1000000)
Y = np.random.uniform(size=1000000)
X_q, Y_q = q_q(X,Y)
plt.scatter(X_q, Y_q)
x = np.linspace(-5,5,100)
f = lambda x:x
plt.plot(x,f(x), '--')
plt.xlabel("Normal distribution quantile")
plt.ylabel("Mesured distribution quantile")
plt.savefig("QQplotuniformnormal.eps")
```
