```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(np.random.randn(1000),np.random.randn(1000),"k*")
plt.plot(np.random.randn(100)+5,np.random.randn(100),"kx")
plt.savefig("anomalies_example.eps")
```

```python
y = np.zeros(800)
y[:300] = np.random.randn(300)
y[300:500] = np.random.randn(200)*2
y[500:] = np.random.randn(300)
plt.plot(y,'k-')
xposition = [300,500]
for xc in xposition:
    plt.axvline(x=xc, color='k', linestyle='--')
plt.savefig("anomalies_variance.eps")
```

```python
y[300:500] = np.random.laplace(size=200, scale=1/np.sqrt(2))
for xc in xposition:
    plt.axvline(x=xc, color='k', linestyle='--')
plt.plot(y,'k-')
plt.savefig("anomalies_laplace.eps")
```

```python
from scipy.stats import norm
from scipy.stats import laplace
```

```python
x = np.linspace(-5,5,1000)
plt.plot(x,norm.pdf(x), label="gaussian_0_1")
plt.plot(x,norm.pdf(x,loc=2), label="gaussian_2_1")
plt.plot(x,norm.pdf(x,scale=4), label="gaussian_0_4")
plt.plot(x,laplace.pdf(x, scale=1/np.sqrt(2)), label="laplacian")
plt.legend()
plt.savefig("anomalies_pdf.eps")
```
