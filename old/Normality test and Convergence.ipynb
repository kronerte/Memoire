```python
from API import *
from scipy.stats import kstest, shapiro
from scipy.stats import norm
```

## Premier exemple

```python
kernel = CCK([gaussian(1),laplacian(1)], [1,1])
law_p = normal(0,1)
law_q = normal(10,1)
```

```python
normality_test(kernel, law_p, law_q, 200,100)[1]
```

```python
n_values = [2,6,10,20,50,100,200,500]
m_values = [2,6,10,20,50,100,200,500]

p_values = np.zeros((len(n_values),len(m_values)))

for i in range(len(n_values)):
    for j in range(len(m_values)):
        plt.figure()
        p_values[i,j] = normality_test(kernel, law_p, law_q, m_values[j],n_values[i])[1]
        plt.title(f"normality test with n={n_values[i]} and m={m_values[j]}; p_value={p_values[i,j]}")
        plt.savefig(f"./normality_test_images/normality test with n={n_values[i]} and m={m_values[j]}.eps")

```

```python
with f as open("normalite_ex1.txt","w"):
    np.save(f, p_values)
```

```python

plt.imshow(p_values)
plt.colorbar()
plt.title("p_values for several values of m and n")
plt.savefig(f"./normality_test_images/valeur_p.eps")
```
