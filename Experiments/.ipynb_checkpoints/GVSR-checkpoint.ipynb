# Sigma simplification Error

```python
omega1 = 1
omega2 = 2

mu_p = 0
sigma_p = 2

mu_q = 0.5
sigma_q = 1

m = 1000
n = 100

alpha = 0.1


law_p = normal(mu_p, sigma_p)
law_q = normal(mu_q, sigma_q)
```

```python
x = np.linspace(0,1,100)
true_fx = np.zeros(100)
false_fx = np.zeros(100)
rest_fx = np.zeros(100)
power = np.zeros(100)
for i in range(100):
    kernel = (x[i]*G(omega1) + (1-x[i])*G(omega2))
    power[i] = kernel.normal_power(mu_p, sigma_p, mu_q, sigma_q, alpha, m , n)
    true_fx[i] = kernel.normal_score(mu_p, sigma_p, mu_q, sigma_q, alpha, m , n)
    false_fx[i] =  np.sqrt(m)*kernel.normal_MMD(mu_p, sigma_p, mu_q, sigma_q)/np.sqrt(kernel.normal_variance(mu_p, sigma_p, mu_q, sigma_q, m, n))
    rest_fx[i] = (np.sqrt(kernel.normal_variance(mu_p, sigma_p, mu_p, sigma_p, m, n)
        /kernel.normal_variance(mu_p, sigma_p, mu_q, sigma_q, m, n))*stats.norm.ppf(1-alpha))
```

```python
plt.figure(figsize=(15,10))
plt.subplot(221)
plt.title("True score of the oMMD Test")
plt.plot(x,true_fx,'k')
plt.subplot(222)
plt.title("Wrong score of the oMMD Test")
plt.plot(x,false_fx,'k')
plt.subplot(223)
plt.title("Error score of the oMMD Test")
plt.plot(x,rest_fx,'k')
plt.subplot(224)
plt.title("Power score of the oMMD Test")
plt.plot(x,power,'k')


plt.savefig("error_score.eps")
```

```python
from Anomaly.estimators import meanEmbeding_MC, moyenneMMD_MC
from Anomaly.utils import inv_phi
from Anomaly.MMD import OMMD
from scipy import optimize
```

```python
def partial_covarianceOMMD_MC(Kernels, law_p, law_q, size_gen = 100, finalSampleSize=1000, verbose = 1):
    sample = np.zeros((len(Kernels), finalSampleSize))
    for i in range(finalSampleSize):
        if verbose > 0:
            print(i/finalSampleSize*100, '% achevé',end='\r',flush=True)
        x1 = law_p(1)
        x2 = law_p(1)
        for j in range(len(Kernels)):
            sample[j,i] = Kernels[j](x1,x2) -  meanEmbeding_MC(Kernels[j], x1, law_q, size_gen = size_gen) -  meanEmbeding_MC(Kernels[j], x2, law_q, size_gen = size_gen)
    return np.cov(sample)
```

```python
def cov_OMMD_MC(Kernels, law_p, law_q, size_gen = 100,Lambda = 0.1, finalSampleSize=1000, verbose = 1):
    if verbose > 0:
        print("Start computing first componant")
    first_componant = partial_covarianceOMMD_MC(Kernels, law_p, law_q, size_gen = size_gen, finalSampleSize=finalSampleSize, verbose = verbose)
    if verbose > 0:
        print("Start computing second componant")
    second_componant = partial_covarianceOMMD_MC(Kernels, law_q, law_p, size_gen = size_gen, finalSampleSize=finalSampleSize, verbose = verbose)
    return 2*(first_componant + 1/Lambda*second_componant)
```

```python
def penalite(beta, rho, Moy, Snum, Sden):
    beta = beta.reshape((-1,1))
    return (np.sqrt(beta.T.dot(Snum).dot(beta))*inv_phi(1-alpha) 
            - np.sqrt(m)*Moy.T.dot(beta) 
            + 1/(2*rho)*(beta.T.dot(Sden).dot(beta)-1)**2).flatten()
```

```python
def Dpen(beta, rho, Moy, Snum, Sden):
    beta = beta.reshape((-1,1))
    return (Snum.dot(beta)/np.sqrt(beta.T.dot(Snum).dot(beta))*inv_phi(1-alpha) 
            - np.sqrt(m)*Moy
           + 1/rho*(beta.T.dot(Sden).dot(beta)-1)*Sden.dot(beta)).flatten()
```

```python
def optim_p(penalite, Dpen, x0, methode,Moy, Snum, Sden, eps=1e-10):
    x_old = x0
    rho = 0.00001
    n = x_old.shape[0]
    bounds = [(0,None) for i in range(n)]
    bounds[0] = (0.00000001,None)
    x = methode(lambda x: penalite(x, rho, Moy, Snum, Sden), x_old,
                jac = lambda x: Dpen(x, rho, Moy, Snum, Sden),bounds=bounds)['x']
    L_x = [x_old, x]
    while np.linalg.norm(x_old - x) > eps:
        print("j")
        x_old = x
        rho *= 0.5
        x = methode(lambda x: penalite(x, rho, Moy, Snum, Sden), x_old,
                    jac = lambda x: Dpen(x, rho, Moy, Snum, Sden),bounds=bounds)['x']
        L_x.append(x)
    return x, L_x
```

```python
def AlgOptBar(Kernels, law_p, law_q, m=1000, n=1000, alpha=0.05):
    #Initialisation:
    Sden = cov_OMMD_MC(Kernels, law_p, law_q, finalSampleSize=1000, verbose = 0)
    Snum = cov_OMMD_MC(Kernels, law_p, law_p,finalSampleSize=1000, verbose = 0)
    Moy = np.r_[[moyenneMMD_MC(OMMD, kernel, law_p, law_q, 1000, 100,repeat=1000, verbose=0) for kernel in Kernels]].reshape(-1,1)
    print(Sden)
    print(Snum)
    print(Moy)
    beta = np.ones(len(Kernels))
    beta_optim, beta_hist = optim_p(penalite, Dpen, beta, optimize.minimize,Moy, Snum, Sden)
    beta_optim = beta_optim/np.sum(beta_optim)
    
    return beta_optim, beta_hist
```

```python
def AlgOptBar2(Kernels, law_p, law_q, m=1000, n=100, alpha=0.05):
    #Initialisation:
    
    
    beta = np.ones(len(Kernels))
    beta2 = np.ones((len(Kernels),1))
    K = beta2.T.dot(Kernels.reshape((-1,1)))[0][0]
    
    Sden = 2*K.normal_partial_variance_Matrix(mu_p, sigma_p, mu_q, sigma_q) + 2*m/n*K.normal_partial_variance_Matrix(mu_q, sigma_q, mu_p, sigma_p)
    Snum = 2*(1+m/n)*K.normal_partial_variance_Matrix(mu_p, sigma_p, mu_p, sigma_p)
    Moy = K.normal_vect_MMD(mu_p, sigma_p, mu_q, sigma_q ).reshape((-1,1))
    print(Sden)
    print(Snum)
    print(Moy)
    beta_optim = optim_p(penalite, Dpen, beta, optimize.minimize,Moy, Snum, Sden)
    beta_optim = beta_optim/np.sum(beta_optim)
    
    
    
    return beta_optim
```
