## Characteristic Kernels

```python
x = np.r_[-3:3:1000j].reshape((1000,1))
for sigma in [0.2,0.5,1,2,4]:
    plt.plot(x,Anomaly.kernels.gaussian(sigma)(x,0),label=f"sigma={sigma}")
plt.legend()
plt.title("Gaussian kernel")
#plt.savefig("gaussian_kernel.eps")
```

```python
x = np.r_[-3:3:1000j].reshape((1000,1))
for sigma in [0.2,0.5,1,2,4]:
    plt.plot(Anomaly.kernels.laplacian(sigma)(x,0),label=f"beta={sigma}")
plt.legend()
plt.title("laplacian kernel")
#plt.savefig("laplacian_kernel.eps")
```

```python
for t in [0,0.1,0.3,0.5,0.7,0.9,1]:
    kernel = Anomaly.kernels.CCK([Anomaly.kernels.gaussian(1),Anomaly.kernels.laplacian(1)],[t,1-t])
    plt.plot(kernel(x,0),label=f"t={t}")
plt.legend()
plt.title("Convex Combination")
#plt.savefig("convex.eps")
```

## MMD

```python
law_p = Anomaly.utils.normal(0,1)
law_q = Anomaly.utils.normal(1,1)
```

```python
kdeplot(law_p(10000).ravel(),label = "distribution under normal conditions")
kdeplot(law_q(10000).ravel(),label = "distribution under anormal conditions")
```

```python
MMD = Anomaly.MMD.QMMD
kernel = Anomaly.kernels.gaussian(1)

%time sample = Anomaly.estimators.sampleMMD_MC(MMD, kernel, law_p, law_q, 100, 100, finalSampleSize=1000)

kdeplot(sample,label = "distribution Quadratic MMD Estimator")
```

```python
MMD = Anomaly.MMD.LMMD
kernel = Anomaly.kernels.gaussian(1)

%time sample = Anomaly.estimators.sampleMMD_MC(MMD, kernel, law_p, law_q, 100, 100, finalSampleSize=1000)

kdeplot(sample,label = "distribution Linear MMD Estimator")
```

```python
MMD = Anomaly.MMD.OMMD
kernel = Anomaly.kernels.gaussian(1)

%time sample = Anomaly.estimators.sampleMMD_MC(MMD, kernel, law_p, law_q, 100, 100, finalSampleSize=1000)

kdeplot(sample,label = "distribution Online MMD Estimator")
```
