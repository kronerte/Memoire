### Kernel

```python
gaussian = lambda sigma: lambda x,y : np.exp(-np.linalg.norm(x-y,axis=1)**2/(2*sigma))
laplacian = lambda sigma: lambda x,y : np.exp(-sigma*np.linalg.norm(x-y,axis=1,ord=1))
```

### Probability law

```python
normal = lambda mu, sigma : lambda size : sigma*np.random.randn(size).reshape(-1,1) + mu
uniform = lambda a, b: lambda size : np.random.uniform(low=a, high=b, size=(size,1))
```

### MMDl

```python
class MMDl():
    def __init__(self,kernel,X):
        self.X = X
        self.Y = []
        self.m = X.shape[0]
        self.n = 0
        self.kernel = kernel
        
        X_, Y_ = X[0::2,:],X[1::2,:]
        self.estim_normP = 2/self.m*self.kernel(X_,Y_).sum()
        
        self.sumNormQ = 0
        self.sumPS = 0
        
        self.MMD = self.estim_normP 
           
    def update(self,y1,y2):
        self.sumNormQ += self.kernel(y1,y2)
        
        self.sumPS += self.kernel(self.X,y1).sum() + self.kernel(self.X,y2).sum()
        
        self.n += 2
        
        self.MMD = self.estim_normP + 2/self.n * self.sumNormQ - 2/(self.n*self.m)*self.sumPS
        
    def fit(self, Y):
        for i in range(0,Y.shape[0]//2):
            self.update(Y[2*i].reshape((-1,1)),Y[2*i+1].reshape((-1,1)))
```

### Estimateurs

```python
def moyenne(kernel, X_, Y, repeat=100):
    mmd = np.zeros(repeat)
    for i in range(repeat):
        MMD = MMDl(kernel,np.random.choice(X_.flatten(), size = 1000, replace=True).reshape((-1,1)))
        MMD.fit(np.random.choice(Y.flatten(), size = 100, replace=True).reshape((-1,1)))
        mmd[i] = MMD.MMD
        #print(f"Progression {(i/repeat)*100:.0f}%", end="\r", flush=True)
    return mmd.mean()
```

```python
kernel = gaussian(1)
law_p = normal(0,1)
law_q = normal(0,1)
X = law_p(10000)
Y = law_q(10000)
moyenne(kernel, X, Y,100)
```

```python
def est_partial_mean(kernel, X, y):
    return np.mean(kernel(y, X))
```

```python
est_partial_mean(gaussian(1),normal(0,1)(1000),10)
```

```python
def estim_V(kernel, X, Y, repeat=100):
    echantillions = np.zeros(repeat)
    for i in range(repeat):
        x1 = np.random.choice(X.flatten(), size = 1, replace=True).reshape((-1,1))
        x2 = np.random.choice(X.flatten(), size = 1, replace=True).reshape((-1,1))
        Y_1 = np.random.choice(Y.flatten(), size = 1000, replace=True).reshape((-1,1))
        Y_2 = np.random.choice(Y.flatten(), size = 1000, replace=True).reshape((-1,1))
        echantillions[i] = kernel(x1,x2) - est_partial_mean(kernel, x1, Y_1) - est_partial_mean(kernel, x2, Y_2)
    return echantillions.var()
    
```

```python
def var_H(kernel, X, Y, alpha, size=1000):
    return estim_V(kernel, X, Y) + 1/alpha*estim_V(kernel, Y, X)
```

```python
def sampleMMD(kernel, law_p, law_q, m, n):
    repeat = 1000
    Liste_mmdl = []
    for _ in range(repeat):
        if (_%10==0):
            print(_/repeat*100, '% achevé',end='\r',flush=True)
        X = law_p(m)
        Y = law_q(n)

        MMD =  MMDl(kernel, X)
        MMD.fit(Y)
        Liste_mmdl.append(MMD.MMD[0])
    return np.r_[Liste_mmdl]
```

## Treshold

```python
norm.ppf(1-0.05)
```

```python
def compute_treshold(kernel, m, n, law_p, alpha):
    X = law_p(10000)
    sigma_H0 = np.sqrt(var_H(kernel, X, X, n/m, size=1000))
    return np.sqrt(2)*sigma_H0*norm.ppf(1-alpha)/np.sqrt(m)
    
```

```python
th = compute_treshold(kernel, m, n, law_p, 0.05)
```

```python
L = sampleMMD(kernel, law_p, law_p, m, n)
```

```python
plt.plot(L,"*")
plt.plot(np.ones(1000)*th)
```

```python
(L > th).mean()
```

```python
def objectif(kernel, n, m, law_p, law_q, alpha):
    """ A minimiser"""
    th = compute_treshold(kernel, n, m, law_p, alpha)
    X = law_p(10000)
    Y = law_q(10000)
    sigma_H1 = np.sqrt(var_H(kernel, X, Y, n/m, size=1000))
    moy = moyenne(kernel, X, Y, repeat=1000)
    
    return np.sqrt(m/2)*(th-moy)/sigma_H1
```

```python
objectif(kernel, n, m, law_p, law_q, 0.05)
```

```python
parameter = [0.5,1,2,4,10,20,40,50,1000]
obj = []
for p in parameter:
    kernel = gaussian(p)
    m = 1000
    n = 100
    law_p = normal(0,1)
    law_q = normal(1,1)
    obj.append(objectif(kernel, n, m, law_p, law_q, 0.05))
```

```python
plt.plot(parameter, obj)
plt.title("obj sur un noyau gaussien")
```

# Calcul

```python
dataset = pd.DataFrame(columns=["Kernel","Kernel_parameter","lambda","law_p","law_q","sigma_H0","sigma_H1","moyenne"])


Kernel = ['gaussian','laplacian']
Ker = {}
Ker['gaussian'] = gaussian
Ker['laplacian'] = laplacian

parametre = [0.25, 0.5, 1, 2, 5]

lambda_ = [1, 0.8, 0.5, 0.2, 0.1, 0.05, 0.01, 0.001]

Law_p = ["normal_0_1", "unif_-1_1"]
law_p = {}
law_p["normal_0_1"] = normal(0,1)
law_p["unif_-1_1"] = uniform(-1,1)


Law_q = ["normal_1_1","normal_0_2", "unif_0_1"]
law_q = {}
law_q["normal_1_1"] = normal(1,1)
law_q["normal_0_2"] = normal(0,2)
law_q["unif_0_1"] = uniform(0,1)




n = 0
for kern in Kernel:
    for p in parametre:
        for l in lambda_:
            for l_p in Law_p:
                for l_q in Law_q:
                    
                    X = law_p[l_p](10000)
                    Y = law_q[l_q](10000)
                    
                    sigma_h0 = np.sqrt(var_H(Ker[kern](p), X, X, l, size=500))
                    sigma_h1 = np.sqrt(var_H(Ker[kern](p), X, Y, l, size=500))
                    moy = moyenne(Ker[kern](p), X, Y, repeat=500)
                    
                    dataset.loc[n] = [kern, p,l,l_p,l_q,sigma_h0,sigma_h1,moy]
                    
                    n += 1
                    
                    print( f"{100*n/480:.2f}%", flush=True, end="\r")

        

dataset.to_csv("optimisation_data.csv", encoding='utf-8', index=False)  
```

Moyenne

```python
def moyenne(kernel, X_, Y, repeat=100):
    mmd = np.zeros(repeat)
    for i in range(repeat):
        MMD = MMDl(kernel,np.random.choice(X_.flatten(), size = 100, replace=True).reshape((-1,1)))
        MMD.fit(np.random.choice(Y.flatten(), size = 100, replace=True).reshape((-1,1)))
        mmd[i] = MMD.MMD
    return mmd.mean()
```

```python
dataset = pd.DataFrame(columns=["Kernel","Kernel_parameter","lambda","law_p","law_q","moyenne"])


Kernel = ['gaussian','laplacian']
Ker = {}
Ker['gaussian'] = gaussian
Ker['laplacian'] = laplacian

parametre = [0.25, 0.5, 1, 2, 5]

lambda_ = [1, 0.8, 0.5, 0.2, 0.1, 0.05, 0.01, 0.001]

Law_p = ["normal_0_1", "unif_-1_1"]
law_p = {}
law_p["normal_0_1"] = normal(0,1)
law_p["unif_-1_1"] = uniform(-1,1)


Law_q = ["normal_1_1","normal_0_2", "unif_0_1"]
law_q = {}
law_q["normal_1_1"] = normal(1,1)
law_q["normal_0_2"] = normal(0,2)
law_q["unif_0_1"] = uniform(0,1)




n = 0
for kern in Kernel:
    for p in parametre:
        for l in lambda_:
            for l_p in Law_p:
                for l_q in Law_q:
                    for _ in range(10):
                        X = law_p[l_p](1000)
                        Y = law_q[l_q](1000)


                        moy = moyenne(Ker[kern](p), X, Y, repeat=100)

                        dataset.loc[n] = [kern, p,l,l_p,l_q,moy]

                        n += 1

                        print( f"{100*n/4800:.2f}%", flush=True, end="\r")

        

dataset.to_csv("moyenne_data.csv", encoding='utf-8', index=False) 
```

### Calcul du terme 1

```python
inv_phi = lambda x : norm.ppf(x)


def V1V2(sigma_H0, alpha, sigma_H1):
    """ Renvoie le premier terme de l'élément à optimiser"""
    return sigma_H1*inv_phi(1 - alpha)/sigma_H0
```

```python
data_set["V1V2_alpha05"] = V1V2(data_set["sigma_H0"],0.05,data_set["sigma_H1"])
```

```python
@interact(Kernel = ['gaussian','laplacian'],
          Lambda = [1, 0.8, 0.5, 0.2, 0.1, 0.05, 0.01, 0.001], 
         Law_p = ["normal_0_1", "unif_-1_1"],
         Law_q = ["normal_1_1","normal_0_2", "unif_0_1"])
def f(Kernel,Lambda,Law_p,Law_q):
    data_set[data_set['Kernel']==Kernel
        ][data_set['lambda']==Lambda
         ][data_set['law_p']==Law_p
          ][data_set['law_q']==Law_q].plot(y="V1V2_alpha05",x='Kernel_parameter')
```

### Calcul du terme 2

```python
def MV(m, MMD_H1, sigma_H1):
    """ Renvoie le deuxième terme de l'élément à optimiser"""
    return np.sqrt(m/2)*MMD_H1/sigma_H1
```

```python
data_set["MV_m1000"] = MV(1000, data_set["moyenne"], data_set["sigma_H1"])
```

# Recherche dynamique

```python
def compute_opt(law_H0, law_H1, Lambda, kernel, kernel_params, m, alpha):
    size = len(kernel_params)
    sigma_H0 = np.zeros(size)
    sigma_H1 = np.zeros(size)
    moyenne_H1 = np.zeros(size)
    V1V0 = np.zeros(size)
    MV_ = np.zeros(size)
    opt = np.zeros(size)
    for i in range(size):
        param = kernel_params[i]
        X = law_H0(10000)
        Y = law_H1(10000)
                    
        sigma_H0[i] = np.sqrt(var_H(kernel(param), X, X, Lambda, size=500))
        sigma_H1[i] = np.sqrt(var_H(kernel(param), X, Y, Lambda, size=500))
        moyenne_H1[i] = moyenne(kernel(param), X, Y, repeat=500)
        
    V1V0 = V1V2(sigma_H0, alpha, sigma_H1)
    MV_ = MV(m, moyenne_H1, sigma_H1)
    
    opt = V1V0 - MV_
    
    return sigma_H0, sigma_H1, moyenne_H1, V1V0, MV_, opt
        
        

    
```

```python
law_H0 = normal(0,1)
law_H1 = normal(1,1)
Lambda = 0.1
kernel = gaussian
m = 10000
alpha = 0.05
kernel_params = [0.01, 0.2,1,2,5,10]


sigma_H0, sigma_H1, moyenne_H1, V1V0, MV_, opt = compute_opt(law_H0, law_H1, Lambda, kernel, kernel_params, m, alpha)
```

```python

plt.plot(kernel_params, sigma_H0, label="sigma_H0")
plt.plot(kernel_params, sigma_H1, label="sigma_H1")
plt.plot(kernel_params, moyenne_H1, label="moyenne_H1")
plt.plot(kernel_params, V1V0, label="V1V0")
plt.plot(kernel_params, MV_, label="MV")
plt.plot(kernel_params, opt, label="opt")
plt.legend()
```
