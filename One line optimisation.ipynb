```python
kernel_params = np.linspace(100,10000,10)
kernel = gaussian
law_p = normal(0,1)
law_q = normal(-1,1)
data = np.zeros((10,6,10))
for i in range(10):
    data[i] = np.array(line_optimisation(law_p,
                                         law_q,
                                         Lambda=0.1,
                                         kernel=kernel,
                                         kernel_params=kernel_params,
                                         m=1000, 
                                         alpha=0.05))
```

```python
for i in range(10):
    plt.plot(data[i][0])
plt.title('sigma_h0')
plt.figure()
for i in range(10):
    plt.plot(data[i][1])
plt.title('sigma_h1')
plt.figure()
for i in range(10):
    plt.plot(data[i][2])
plt.title('moyenne')
plt.figure()

plt.plot(data[:,3].mean(axis=0))
plt.title('V1V0')

plt.figure()

plt.plot(data[:,4].mean(axis=0))
plt.title('MV')

plt.figure()

plt.plot(data[:,5].mean(axis=0))
plt.title('puis')
```
